```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
summary_city_df = pd.read_csv('data/car_sales/summary_all_city.csv')
summary_city_df.head(10)
```

```python
x = summary_city_df['City'].head(10)
```

```python
data = summary_city_df['Average Price'].head(10)
```

```python
plt.bar(x,data); plt.xticks(x, ('Atl','Chi','Hou','Mia','Minn','Nash','NY','SA', 'Sea','SF'))
plt.xticks(rotation=0)
plt.xlabel("City")
plt.title("Average Price Per City")
plt.ylabel("Average Price")
```
